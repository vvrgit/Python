```python
roll=int(input("Enter roll no."))
failed_sub=data[data[roll]<35].index
print("Telugu:",data.loc["Tel"].loc[roll])
print("English:",data.loc["Eng"].loc[roll])
print("Hindi:",data.loc["Hin"].loc[roll])
print("Total Marks:",data[roll].sum())
print("Total Marks:",data[roll].max())
print("Total Marks:",data[roll].min())
print("Total Marks:",data[roll].mean())
print("Failed Subjects:",failed_sub.values)
```
